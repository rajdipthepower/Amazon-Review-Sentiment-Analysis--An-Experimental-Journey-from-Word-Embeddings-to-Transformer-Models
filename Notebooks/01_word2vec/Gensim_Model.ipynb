```python
from gensim.models import Word2Vec
from gensim.utils import simple_preprocess
import pandas as pd
```

```python
from pickle import TRUE
df = pd.read_json("Sports_and_Outdoors_5.json",lines=True)
df.head()
```

```python
review_text_tokenized = df['reviewText'].apply(simple_preprocess)
```

```python
review_text_tokenized
```

```python
model = Word2Vec(vector_size=200,min_count = 5 , workers=4 , window = 5)

# min_count is freuency limit , hence words with freq<5 will be ignored
#window is the sliding window
```

```python
model.build_vocab(review_text_tokenized,progress_per=10000)
model.train(review_text_tokenized,total_examples = model.corpus_count , epochs = 15)
```

```python
model.wv.most_similar('awful')
```

```python
print(model.wv.similarity('good','great'))
print(model.wv.similarity('slow','steady'))
```
